```python
import pandas as pd
```

```python
path = 'data_preprocessing/식품영양성분DB_전처리.csv'
df = pd.read_csv(path, encoding='utf-8', low_memory=False)
```

```python
df.shape
```

```python
for i, j in zip(df.columns, df.dtypes):
    print(f'{i}: {j}')
```

```python
# AMT_NUM22 고유값 및 데이터 타입 출력
print("AMT_NUM22 고유값 및 데이터 타입:")
unique_vals = df['AMT_NUM22'].dropna().unique()

for val in unique_vals:
    print(f"값: {val!r}, 타입: {type(val)}")
```
