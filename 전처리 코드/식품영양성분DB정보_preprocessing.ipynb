### 전처리

```python
import pandas as pd

# ----------------------------
# 1. CSV 로드
# ----------------------------
df = pd.read_csv('../data/식품영양성분DB정보.csv', encoding='utf-8-sig', low_memory=False)

# ----------------------------
# 2. AMT_NUM 컬럼 중 전체 결측치인 컬럼 제거
# ----------------------------
amt_cols = [col for col in df.columns if col.startswith('AMT_NUM')]
amt_cols_all_nan = [col for col in amt_cols if df[col].isna().all()]
df = df.drop(columns=amt_cols_all_nan)

# ----------------------------
# 3. 주요 영양소 결측 행 제거 (에너지, 단백질, 지방, 탄수화물)
# ----------------------------
key_cols = ['AMT_NUM1', 'AMT_NUM3', 'AMT_NUM4', 'AMT_NUM6']
existing_key_cols = [col for col in key_cols if col in df.columns]
df = df.dropna(subset=existing_key_cols, how='any')

# ----------------------------
# 4. 주요 영양소 컬럼명을 한글로 변경
# ----------------------------
rename_map = {
    'AMT_NUM1': '에너지(kcal)',
    'AMT_NUM3': '단백질(g)',
    'AMT_NUM4': '지방(g)',
    'AMT_NUM6': '탄수화물(g)',
    'AMT_NUM7': '당류(g)',
    'AMT_NUM13': '나트륨(mg)',
    'AMT_NUM23': '콜레스테롤(mg)',
    'AMT_NUM24': '포화지방산(g)',
    'AMT_NUM25': '트랜스지방산(g)',
}
df = df.rename(columns={k: v for k, v in rename_map.items() if k in df.columns})

# ----------------------------
# 5. AMT_NUM 계열 → 모두 float 강제 변환
# ----------------------------
amt_cols_existing = [col for col in amt_cols if col in df.columns]
df[amt_cols_existing] = df[amt_cols_existing].apply(pd.to_numeric, errors='coerce')

# 주요 영양소 한글 컬럼도 float 변환 (중복 방지 위해 필터링)
numeric_renamed_cols = [v for k, v in rename_map.items() if v in df.columns]
df[numeric_renamed_cols] = df[numeric_renamed_cols].apply(pd.to_numeric, errors='coerce')


# ----------------------------
# 6. 불필요 컬럼 제거
# ----------------------------
drop_cols = ['Z10500', 'NATION_NM', 'IMP_YN', 'ITEM_REPORT_NO', 'MAKER_NM', 'UPDATE_DATE','SERVING_SIZE']
df = df.drop(columns=[col for col in drop_cols if col in df.columns])

# ----------------------------
# 7. 중복 식품 제거
# ----------------------------
df = df.drop_duplicates(subset='FOOD_NM_KR', keep='first')

# ----------------------------
# 8. 저장
# ----------------------------
df.to_csv('../data_preprocessing/식품영양성분DB_전처리.csv', index=False, encoding='utf-8-sig')
print(" 전처리 완료: ../data_preprocessing/식품영양성분DB_전처리.csv 저장됨")
```
