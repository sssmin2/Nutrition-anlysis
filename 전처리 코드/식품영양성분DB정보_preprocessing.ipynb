### 전처리

```python
import pandas as pd

# ----------------------------
# 1. CSV 로드 (한글 인코딩 주의)
# ----------------------------
df = pd.read_csv('../data/식품영양성분DB정보.csv', encoding='utf-8-sig', low_memory=False)

# ----------------------------
# 2. 추출할 컬럼 정의 및 컬럼명 매핑
# ----------------------------
col_map = {
    'FOOD_CD': 'food_code',
    'FOOD_NM_KR': 'food_name',
    'SERVING_SIZE': 'serving_size(g)',
    'Z10500':'net_weight(g)',
    'AMT_NUM1': 'calorie_kcal',
    'AMT_NUM3': 'protein_g',
    'AMT_NUM4': 'fat_g',
    'AMT_NUM6': 'carbohydrate_g',
    'AMT_NUM7': 'sugar_g',
    'AMT_NUM13': 'sodium_mg',
    'AMT_NUM14': 'vitamin_a_μg',
    'AMT_NUM21': 'vitamin_c_mg',
    'AMT_NUM22': 'vitamin_d_mg',
    'AMT_NUM23': 'cholesterol_mg',
    'AMT_NUM24': 'saturated_fat_g',
    'AMT_NUM25': 'trans_fat_g',
    'AMT_NUM30': 'vitamin_b12_μg',
    'AMT_NUM116': 'zinc_mg',
    
}

df_nutr = df[list(col_map.keys())].rename(columns=col_map)
df_nutr['serving_size(g)'] = df_nutr['serving_size(g)'].str.extract(r'(\d+\.?\d*)').astype(float)
df_nutr['net_weight(g)'] = df_nutr['net_weight(g)'].str.extract(r'(\d+\.?\d*)').astype(float)


# ----------------------------
# 3. 숫자형 컬럼 변환 (문자 → float64)
# ----------------------------
numeric_cols = list(col_map.values())[2:]  # serving_size부터 끝까지
df_nutr[numeric_cols] = df_nutr[numeric_cols].apply(pd.to_numeric, errors='coerce')

# ----------------------------
# 4. 주요 성분 결측 제거
# ----------------------------
key_cols = ['calorie_kcal', 'carbohydrate_g', 'protein_g', 'fat_g']
df_nutr = df_nutr.dropna(subset=key_cols, how='any')

# ----------------------------
# 5. 중복 제거 (식품명 기준)
# ----------------------------
df_nutr = df_nutr.drop_duplicates(subset='food_name')

# ----------------------------
# 6. 저장
# ----------------------------
df_nutr.to_csv('../data_preprocessing/식품영양성분DB_전처리.csv', index=False, encoding='utf-8-sig')
print("전처리 완료: '../data_preprocessing/식품영양성분DB_전처리.csv'")
```
