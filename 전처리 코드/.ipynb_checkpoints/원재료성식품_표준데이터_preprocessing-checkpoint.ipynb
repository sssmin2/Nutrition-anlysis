#### 전처리

```python
import pandas as pd
import re

# ----------------------------
# 1. CSV 로드 
# ----------------------------
path = '../data/전국통합식품영양성분정보_원재료성식품_표준데이터.csv'
df = pd.read_csv(path, encoding='cp949', low_memory=False)

# ----------------------------
# 2. 주요 성분만 추출 (A 전략 기준)
# ----------------------------
nutrient_cols_map = {
    '식품코드': 'food_code',
    '식품명': 'food_name',
    '데이터구분코드': 'data_pid',
    '데이터구분명': 'data_pid_name',
    '영양성분함량기준량': 'serving_size(g)',
    '에너지(kcal)': 'calorie_kcal',
    '탄수화물(g)': 'carbohydrate_g',
    '단백질(g)': 'protein_g',
    '지방(g)': 'fat_g',
    '당류(g)': 'sugar_g',
    '포화지방산(g)': 'saturated_fat_g',
    '트랜스지방산(g)': 'trans_fat_g',
    '나트륨(mg)': 'sodium_mg',
    '콜레스테롤(mg)': 'cholesterol_mg',
    '식이섬유(g)': 'dietary_fiber_g',
    '비타민 A(μg RAE)': 'vitamin_a_μg',
    '비타민 C(mg)': 'vitamin_c_mg',
    '비타민 D(μg)': 'vitamin_d_ug',
    '칼슘(mg)': 'calcium_mg',
    '철(mg)': 'iron_mg',
}

df_nutr = df[list(nutrient_cols_map.keys())].rename(columns=nutrient_cols_map)
df_nutr['serving_size(g)'] = df_nutr['serving_size(g)'].str.extract(r'(\d+\.?\d*)').astype(float)

# ----------------------------
# 3. 숫자형 컬럼 변환 (문자 → float64)
# ----------------------------
numeric_cols = list(nutrient_cols_map.values())[4:]  # serving_size부터 끝까지
df_nutr[numeric_cols] = df_nutr[numeric_cols].apply(pd.to_numeric, errors='coerce')

# ----------------------------
# 4. 주요 성분 결측 제거
# ----------------------------
key_cols = ['calorie_kcal', 'carbohydrate_g', 'protein_g', 'fat_g']
df_nutr = df_nutr.dropna(subset=key_cols, how='any')

# ----------------------------
# 5. 중복 제거 (식품명 기준)
# ----------------------------
df_nutr = df_nutr.drop_duplicates(subset='food_name')

# ----------------------------
# 7. 저장
# ----------------------------
df_nutr.to_csv('../data_preprocessing/원재료성식품_전처리.csv', index=False, encoding='utf-8-sig')
print("전처리 완료: 원재료성식품_전처리.csv 저장됨")
```
